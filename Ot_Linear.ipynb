### Otimização Linear - Gabriel Haeser

```python
# Imports e tratamento dos dados

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv('Notas_P2_Ot_Linear.csv',sep=';')
df.index = df['Código']
df = df.drop(['Código'], axis=1)

to_drop = df[(df['Prova1'] == 0) | (df['Prova2'] == 0)].index
df_no_zeros = df.drop(to_drop)
```

Histograma das P1 e P2

```python
df_no_zeros.hist(figsize=(22,5))
plt.show()
```

Plot KDE da P1 e P2, excluindo-se as pessoas que zeraram (ou não fizeram) alguma das provas.

```python
plt.figure(figsize=(20,5))
sns.kdeplot(df_no_zeros[df.columns[0]], shade=True)
sns.kdeplot(df_no_zeros[df.columns[1]], shade=True)
plt.xlabel('Notas')
plt.ylabel('Freq')
plt.show()
```

```python
for prova in [0, 1]:
    print('Prova {}'.format(prova+1))
    print('Média:', df_no_zeros.mean()[prova])
    print('Mediana:', df_no_zeros.median()[prova])
    print('Desvio padrão:', df_no_zeros.std()[prova], '\n')
```

___

Histograma da média das provas (excluindo-se os que zeraram alguma das provas)

```python
df['Média P1 e P2'] = df.apply(lambda row : (row['Prova1'] + row['Prova2']) / 2, axis=1)
df_no_zeros['Média P1 e P2'] = df_no_zeros.apply(lambda row : (row['Prova1'] + row['Prova2']) / 2, axis=1)

df_no_zeros[['Média P1 e P2']].hist(figsize=(22,5), normed=True)
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.xticks(np.arange(0,11,1))
plt.show()
```

```python
print('Média (P1 e P2):', np.mean(df_no_zeros['Média P1 e P2']))
print('Mediana (P1 e P2):', np.median(df_no_zeros['Média P1 e P2']))
print('Des. Padrão:', np.std(df_no_zeros['Média P1 e P2']))
```

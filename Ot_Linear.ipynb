```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Notas_P2_Ot_Linear.csv',sep=';')
df.index = df['Código']
df = df.drop(['Código'], axis=1)

to_drop = df[(df['Prova1'] == 0) | (df['Prova2'] == 0)].index
df_no_zeros = df.drop(to_drop)
```

```python
plt.figure(figsize=(20,5))
sns.kdeplot(df_no_zeros[df.columns[0]], shade=True)
sns.kdeplot(df_no_zeros[df.columns[1]], shade=True)
plt.xlabel('Notas')
plt.ylabel('Freq')
plt.show()
```

```python
plt.figure(figsize=(22,5))
sns.distplot(df_no_zeros[['Prova1']], hist=True, kde=False)
sns.distplot(df_no_zeros[['Prova2']], hist=True, kde=False)
plt.title('P1 : Azul, P2 : Laranja')
plt.xlabel('Notas')
plt.ylabel('Freq')
# plt.xlim([-5, 13])
plt.show()
```

```python
plt.figure(figsize=(22,5))
sns.distplot(df_no_zeros[['Prova1']], hist=True, kde=False)
sns.distplot(df_no_zeros[['Prova2']], hist=True, kde=False)
plt.title('P1 : Azul, P2 : Laranja')
plt.xlabel('Notas')
plt.ylabel('Freq')
# plt.xlim([-5, 13])
plt.show()
```

```python
print('Prova 1')
print('Média:', df_no_zeros.mean()[0])
print('Mediana:', df_no_zeros.median()[0])
print('Desvio padrão:', df_no_zeros.std()[0])
```

```python
print('Prova 2')
print('Média:', df_no_zeros.mean()[1])
print('Mediana:', df_no_zeros.median()[1])
print('Desvio padrão:', df_no_zeros.std()[1])
```

```python
df['Média P1 e P2'] = df.apply(lambda row : (row['Prova1'] + row['Prova2']) / 2, axis=1)
df_no_zeros['Média P1 e P2'] = df_no_zeros.apply(lambda row : (row['Prova1'] + row['Prova2']) / 2, axis=1)
```

```python
df[['Média P1 e P2']].hist(figsize=(22,5), normed=True)
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.xticks(np.arange(0,11,1))
plt.show()
```

```python
print('Média (P1 e P2):', np.mean(df_no_zeros['Média P1 e P2']))
print('Mediana (P1 e P2):', np.median(df_no_zeros['Média P1 e P2']))
print('Des. Padrão:', np.std(df_no_zeros['Média P1 e P2']))
```

```python
plt.figure(figsize=(20,5))
plt.title('Média da P1 e P2')
sns.distplot(df_no_zeros[['Média P1 e P2']], kde=True)
plt.xticks(np.arange(0,11,1))
plt.show()
```
